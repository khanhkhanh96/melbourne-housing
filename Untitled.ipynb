# model sử dụng PCA với cross validation dùng k-fold

```python

from sklearn.model_selection import KFold
kf = KFold(n_splits=10, shuffle = False, random_state = 0) #define number of splits
sqerror=[]
for train_idx, test_idx in kf.split(X):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = Y.iloc[train_idx], Y.iloc[test_idx]
    """
    Scale data
    from sklearn.preprocessing import StandardScaler
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)"""

    # Applying PCA
    from sklearn.decomposition import PCA
    pca = PCA()
    X_train = pca.fit_transform(X_train)
    X_test = pca.transform(X_test)
    explained_variance = pca.explained_variance_ratio_

    # Train Regression Model with PCA
    from sklearn.linear_model import LinearRegression
    modelloop = LinearRegression()
    modelloop.fit(X_train, y_train)
    sqthisround=mean_squared_error(y_test, modelloop.predict(X_test))
    sqerror.append( sqthisround ) #append mean square this round
    fig, ax = plt.subplots()
    ax.scatter(y_test, modelloop.predict(X_test))
    ax.plot([Y.min(), Y.max()], [Y.min(), Y.max()], 'k--', lw=4)
    ax.set_xlabel('Measured')
    ax.set_ylabel('Predicted')
    plt.show()
from statistics import mean
print("mean square error of PCA cross validation:", mean(sqerror))
```
